```python
porcentaje_titanic= titanic["Survived"].value_counts()/titanic["Survived"].count()*100
porcentaje_titanic
```

```python
#personas que estabn en segunda clase en orden alfabetico
pasajero_segunda = titanic[titanic["Pclass"]==2]['Name'].sort_values()
pasajero_segunda.head(5)
```

```python
titanic=titanic.dropna(subset=["Age"]) # elimino los na, y datps no regitrados
titanic
```

```python
#porcentaje de personas que murieron en casa clase

porcentaje_no_sobrevivientes = titanic.groupby('Pclass')['Survived'].value_counts()
porcentaje_no_sobrevivientes
titanic_por_clase=titanic.groupby('Pclass')['Survived'].value_counts(normalize=True)*100
titanic_por_clase
```

```python
#edad media de c/hombre y mujer que viejaba por clase
edad_media = titanic.groupby(['Pclass', 'Sex'])['Age'].mean()
edad_media
```
