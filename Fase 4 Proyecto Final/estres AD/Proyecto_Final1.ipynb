# **Acerca de: Las respuestas a las Escalas de Depresión Ansiedad Estrés **

Las respuestas a las Escalas de Depresión Ansiedad Estrés (DASS, por sus siglas en inglés) pueden variar ampliamente según la persona y su situación específica. Estas escalas son herramientas psicométricas diseñadas para medir los niveles de depresión, ansiedad y estrés percibidos por un individuo durante un período de tiempo determinado. Los resultados pueden variar desde niveles bajos hasta niveles severos en cada una de estas dimensiones, y pueden proporcionar información valiosa para el diagnóstico y la planificación del tratamiento. Sin embargo, siempre es importante interpretar los resultados en el contexto individual y considerar otros factores relevantes para la salud mental de la persona.

El conjunto de datos consta de preguntas, respuestas y metadatos recopilados de 39,775 Escalas de Depresión, Ansiedad y Estrés (DASS). Estas escalas son comúnmente utilizadas en evaluaciones psicológicas para medir los niveles de depresión, ansiedad y estrés experimentados por individuos.

Los datos fueron recopilados y alojados en OpenPsychometrics.org, que es una organización sin fines de lucro con el objetivo de educar al público sobre psicología y recopilar datos para fines de investigación psicológica. Información más detallada sobre los datos recopilados, incluidas notas, se puede encontrar en el archivo codebook.txt asociado con el conjunto de datos. Este documento probablemente proporciona contexto adicional sobre las variables medidas, el proceso de administración de la encuesta y cualquier limitación o consideración relevante para interpretar los datos.

**Enfoque del proyecto**

El estrés es una respuesta física y emocional del cuerpo a situaciones que pueden ser percibidas como desafiantes o amenazantes. Puede ser causado por diversas situaciones, como presión laboral, problemas familiares, cambios importantes en la vida, entre otros.

Los síntomas del estrés pueden variar, pero comúnmente incluyen tensión muscular, fatiga, dificultad para concentrarse, irritabilidad y cambios en el estado de ánimo. El estrés prolongado puede tener efectos negativos en la salud física y mental, por lo que es importante aprender a manejarlo de manera efectiva.

Prevalencia del estrés: Según la Organización Mundial de la Salud (OMS), el estrés es considerado como uno de los problemas de salud ocupacional más importantes del siglo XXI. Se estima que el estrés afecta a más del 70% de la población mundial en algún momento de sus vidas.

Impacto en el trabajo: Un estudio realizado por la American Institute of Stress encontró que el estrés laboral es la principal causa de ausentismo en el trabajo, así como de costos médicos y de atención de la salud. Se estima que el estrés laboral cuesta a las empresas estadounidenses más de $300 mil millones anualmente en pérdida de productividad.

Impacto en la salud mental: La Asociación Americana de Psicología informa que el estrés crónico puede contribuir al desarrollo de trastornos de ansiedad y depresión. Además, puede aumentar el riesgo de desarrollar enfermedades cardiovasculares, trastornos gastrointestinales y otros problemas de salud física.

Estrés y juventud: Según datos de la Encuesta Nacional de Salud Infantil de Estados Unidos, aproximadamente el 20% de los adolescentes informan que experimentan niveles elevados de estrés. Factores como la presión académica, las preocupaciones sobre el futuro y los desafíos sociales pueden contribuir a niveles altos de estrés en esta población.

```python
# Filtrar las columnas que comienzan con 'T'
filtered_columns_T = [col for col in data.columns if col.startswith('T')]

# Crear un DataFrame auxiliar con solo estas columnas
filtered_data_T = data[filtered_columns_T]

# Calcular la frecuencia de cada valor en cada columna
frequencies_T = filtered_data_T.apply(pd.Series.value_counts)

# Normalizar los datos
normalized_frequencies_T = frequencies_T.div(frequencies_T.sum(axis=1), axis=0)

# Definir el tamaño del gráfico
plt.figure(figsize=(12, 8))

# Iterar sobre cada columna y crear un gráfico de barras
for i, col in enumerate(normalized_frequencies_T.columns):
    plt.subplot(2, 5, i+1)  # Ajustar el diseño del subplot
    ax = normalized_frequencies_T[col].plot(kind='bar', color='skyblue')  # Crear el gráfico de barras
    plt.title(col)  # Agregar título
    plt.xlabel('Valor')  # Etiqueta del eje x
    plt.ylabel('Frecuencia Normalizada')  # Etiqueta del eje y
    plt.xticks(rotation=0)  # Ajustar la orientación de las etiquetas del eje x
    plt.tight_layout()  # Ajustar el espaciado entre subplots

    # Agregar el valor de frecuencia en cada barra
    for p in ax.patches:
        ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()  # Mostrar los gráficos
```
