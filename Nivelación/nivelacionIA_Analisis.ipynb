```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = '/content/dataset_banco.csv'
data = pd.read_csv(ruta)
```

```python
print(data.shape)
```

```python
data.info()
```

1.- Datos faltantes en alguna celda. Celdas en blanco.
2.- Columnas irrelevantes. Hobbies.
3.- Registros repetidos. Eliminar filas repetidas.
4.- Valores extremos. (outliers). Edades de 200 años.
5.- Erroes tipográficos. Married -> married

```python
data.dropna(inplace=True)#Elimina la fila completa donde encuentre datos faltantes.
data.info()
```

```python
cols_cat = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
for col in cols_cat:
  print(f'columna{col}: {data[col].nunique()} subniveles')
```

```python
data.describe()#Nos deja ver todas las columnas numericas.
```

```python
print(f'Tamaño del set antes de eliminar las filas repetidas: {data.shape}')
data.drop_duplicates(inplace=True)#Elimina las filas duplicadas.
print(f'Tamaño del set despúes de eliminar las filas repetidas: {data.shape}')
```

```python
cols_num =['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8, 30))#Crea una figura de 7 subgraficos dispuesto en una columna.

fig.subplots_adjust(hspace=0.5)#Ajusta el espaciado vértica entre subgraficos

for i, col in enumerate(cols_num):
  sns.boxplot(x=col, data=data, ax=ax[i])
  ax[i].set_title(col)
```

```python
print(f'Tamaño de set antes de depurar la edad: {data.shape}')
data = data[data['age']<=100]
print(f'Tamaño del set despúes de depurar: {data.shape}')
```

```python
print(f'Tamaño de set antes de depurar la duration: {data.shape}')
data=data[data['duration']>0]
print(f'Tamaño de set antes de depurar la duration: {data.shape}')
```
